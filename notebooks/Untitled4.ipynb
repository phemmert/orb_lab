```python
import numpy as np
import pandas as pd
import math
import sys
sys.path.insert(0, r'C:\Users\phemm\orb_lab\src')

def wma(series, length):
    weights = np.arange(1, length + 1)
    return series.rolling(length).apply(
        lambda x: np.dot(x, weights) / weights.sum(), raw=True
    )

def hma(series, length):
    half = length // 2
    sqrt_len = int(math.sqrt(length))
    wma_half = wma(series, half)
    wma_full = wma(series, length)
    raw = 2 * wma_half - wma_full
    return wma(raw, sqrt_len)

# Use backtester to load data (that works)
from orb_backtester import ORBBacktester
bt = ORBBacktester(symbol='AMD', start_date='2025-12-01', end_date='2025-12-01',
                   enable_confluence=False)
bt.load_data()

dec1 = bt.df[bt.df.index.date.astype(str) == '2025-12-01'].copy()

dec1['ssl_baseline_test'] = hma(dec1['close'], 45)
dec1['ssl_gap_test'] = dec1['close'] - dec1['ssl_baseline_test']
dec1['ssl_widening_test'] = dec1['ssl_gap_test'] > dec1['ssl_gap_test'].shift(1)

for t in ['09:34', '09:35', '09:36', '09:37']:
    row = dec1[dec1.index.strftime('%H:%M') == t].iloc[0]
    prev_gap = dec1['ssl_gap_test'].shift(1).loc[row.name]
    print(f"{t}: close={row['close']:.2f} base={row['ssl_baseline_test']:.2f} "
          f"gap={row['ssl_gap_test']:.3f} prev={prev_gap:.3f} widen={row['ssl_widening_test']}")
```

```python
import sys
sys.path.insert(0, r'C:\Users\phemm\orb_lab\src')
import importlib, orb_backtester
importlib.reload(orb_backtester)
from orb_backtester import ORBBacktester
import pandas as pd

bt = ORBBacktester(symbol='AAPL', start_date='2025-12-02', end_date='2025-12-02',
                   enable_confluence=True, confluence_mode='volume', min_confluence=2, verbose=False)
bt.load_data()
df = bt.df
dec2 = df[df.index.date == pd.Timestamp('2025-12-02').date()]

orb_bars = dec2[(dec2.index.hour == 9) & (dec2.index.minute >= 30) & (dec2.index.minute < 35)]
orb_high = orb_bars['high'].max()
orb_low = orb_bars['low'].min()
print(f"ORB: High={orb_high:.4f}, Low={orb_low:.4f}\n")

for t in [36, 51]:
    bar = dec2[(dec2.index.hour == 9) & (dec2.index.minute == t)]
    if not bar.empty:
        b = bar.iloc[0]
        atr = b['atr_rth']
        threshold = atr * 0.1
        
        o, h, l, c = b['open'], b['high'], b['low'], b['close']
        body = abs(c - o)
        rng = h - l
        body_ratio = body / rng if rng > 0 else 0
        
        open_in = (o >= orb_low) and (o <= orb_high)
        above = c > orb_high
        sufficient = c > (orb_high + threshold)
        strong = body_ratio >= 0.5
        
        print(f"09:{t:02d} OHLC: {o:.2f} {h:.2f} {l:.2f} {c:.2f}")
        print(f"  open_in_range: {open_in} (open={o:.2f} in [{orb_low:.2f}, {orb_high:.2f}])")
        print(f"  above_orb:     {above} (close={c:.2f} > orb_high={orb_high:.2f})")
        print(f"  sufficient:    {sufficient} (close={c:.2f} > {orb_high + threshold:.4f})")
        print(f"  strong_body:   {strong} (ratio={body_ratio:.3f} >= 0.5)")
        print(f"  ALL CONDITIONS: {open_in and above and sufficient and strong}\n")
```
