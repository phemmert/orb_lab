```python
import sys
sys.path.insert(0, r'C:\Users\phemm\orb_lab\src')

from orb_backtester import ORBBacktester

print('=' * 60)
print('QUICK OPERATIONAL TEST - December 2025 only')
print('=' * 60)

# Just one month - should be fast
print('\n[1] Testing WITH confluence...')
bt_conf = ORBBacktester(
    symbol='AMD',
    start_date='2025-12-01',
    end_date='2025-12-31',
    enable_confluence=True,
    min_confluence=5,
    verbose=True  # Show each trade
)
r1 = bt_conf.run()
print(f"\n   Trades: {r1['total_trades']}, Total R: {r1['total_r']:+.2f}")

print('\n[2] Testing WITHOUT confluence...')
bt_no = ORBBacktester(
    symbol='AMD',
    start_date='2025-12-01',
    end_date='2025-12-31',
    enable_confluence=False,
    verbose=True
)
r2 = bt_no.run()
print(f"\n   Trades: {r2['total_trades']}, Total R: {r2['total_r']:+.2f}")

print('\n' + '=' * 60)
print('RESULT')
print('=' * 60)
filtered = r2['total_trades'] - r1['total_trades']
print(f"Confluence filtered: {filtered} trades")
print(f"WITH confluence:    {r1['total_trades']} trades, {r1['total_r']:+.2f}R")
print(f"WITHOUT confluence: {r2['total_trades']} trades, {r2['total_r']:+.2f}R")

if r1['total_trades'] < r2['total_trades']:
    print('\n✅ Confluence filtering is ACTIVE')
else:
    print('\n⚠️  No filtering detected - check indicator calculations')
```
