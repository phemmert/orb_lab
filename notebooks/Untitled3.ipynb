```python
import sys
sys.path.insert(0, r'C:\Users\phemm\orb_lab\src')

from orb_backtester import ORBBacktester

# Run with verbose to see what's happening
bt = ORBBacktester(
    symbol='AMD',
    start_date='2025-12-01',
    end_date='2025-12-31',
    enable_confluence=True,
    min_confluence=5,
    verbose=True
)

# Load data and check indicator values
bt.load_data()

# Check a sample of confluence scores
df = bt.df
print("\nSample confluence scores at 9:36 bars (typical entry time):")
print("=" * 70)

bars_936 = df[df.index.strftime('%H:%M') == '09:36'].tail(10)
for idx, row in bars_936.iterrows():
    print(f"{idx.strftime('%Y-%m-%d %H:%M')}: "
          f"SSL_bull={row.get('ssl_score_bull', 'N/A'):.0f} "
          f"WAE_bull={row.get('wae_score_bull', 'N/A'):.0f} "
          f"QQE_bull={row.get('qqe_score_bull', 'N/A'):.0f} "
          f"Vol={row.get('vol_score', 'N/A'):.2f}")
```
